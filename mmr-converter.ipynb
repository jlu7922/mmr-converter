```python
# Split a record based on fixed positions
def split_record(record, positions):
    return [record[start:end] for start, end in positions]  

with open(filepath, 'r') as file:
    lines = file.readlines()

# Process each line and split into fields
data = [split_record(line, positions) for line in lines]

df = pd.DataFrame(data, columns=columns)

# Trim spaces in "Adjustment Reason Code" before converting to numeric
if "Adjustment Reason Code (ARC)" in df.columns:
    df["Adjustment Reason Code (ARC)"] = df["Adjustment Reason Code (ARC)"].str.strip()
    df["Adjustment Reason Code (ARC)"] = pd.to_numeric(df["Adjustment Reason Code (ARC)"], errors='ignore')  # Convert to numeric, ignore invalids


# Convert other columns to numeric
for column in columns_to_convert:
    if column in df.columns and column != "Adjustment Reason Code (ARC)":
        df[column] = pd.to_numeric(df[column], errors='ignore')  # Convert to numeric, ignore invalids

# Save the df to an excel
df.to_excel(filepath + ".xlsx", sheet_name = sheetname, index=False)

print("The file has been successfully converted to", filepath + ".xlsx")
```
